```python
import os
import pandas as pd
```

```python
tweet_count = 500000
text_query = "afgan mülteci"
since_date = "2021-08-15"
until_date = "2022-02-15"

# Using OS library to call CLI commands in Python
os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> afganmulteci.json'.format(tweet_count, since_date, text_query, until_date))

# Reads the json generated from the CLI command above and creates a pandas dataframe
tweets_afghan = pd.read_json('afganmulteci.json', lines=True)

# Displays first 5 entries from dataframe
tweets_afghan
```

```python
tweets_afghans=tweets_afghan[['date','rawContent','id']]
```

```python
tweets_afghans.to_csv('afganmulteci.csv')
```

```python
tweets=pd.read_csv('afganmulteci.csv')
```

```python
tweets.to_excel('afganmulteci.xlsx')
```
